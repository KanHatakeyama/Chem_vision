```python
import json

# Define the file path
file_path = '/data/2024/Chem_vision/250301ft_qwen/dataset/catalogue/実験器具カタログ.json'

# Read the JSON file
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Display the data
#print(data)
```

```python
base_path = 'dataset/catalogue/'
#base_path = '/data/2024/Chem_vision/250301ft_qwen/dataset/catalogue'

for key in data.keys():
    category=data[key]['category'].replace("カテゴリ","")
    data[key]['path'] = base_path +f"{category}/"+ key

# Display the updated data
print(data)
len(data)
```

```python
broken_list_text="""D_1082
D_1089
D_1093
D_1378
D_1384
D_1468
F_815
G_301"""
broken_list=broken_list_text.split("\n")

```

```python
records=[]
for key in data.keys():
    if key.replace(".png","") in broken_list:
        #print(key)
        continue
    record=data[key]
    record['id']=key

    temp={"query": "これは何でしょうか?", "response": f"{record['product_name']}です｡",
          "images": [f"{record['path']}"], "id": record['id']}
    records.append(temp)
```

```python
records
```

```python
out_path="dataset/annotated.jsonl"
with open(out_path, 'w', encoding='utf-8') as file:
    for record in records:
        json.dump(record, file, ensure_ascii=False)
        file.write('\n')
```
